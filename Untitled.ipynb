## Data Selection

```python
books.columns
```

```python
books = books[["book_id", "title","description","genre"]]
```

```python
books
```

```python
books["tags"] = books["description"] + books["genre"]
books
```

```python
books_df = books.drop(columns=['description', 'genre'])
books_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=100, stop_words='english')
```

```python
cv
```

```python
vector = cv.fit_transform(books_df['tags'].values.astype('U')).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
def recommend_books(book):
    index = books_df[books_df['title']==book].index[0]
    distance = sorted(list(enumerate(similarity[index])),reverse=True, key=lambda vector:vector[1])
    for i in distance[1:6]:
        print(books_df.iloc[i[0]].title)
```

```python
recommend_books('Sharp Objects')
```
